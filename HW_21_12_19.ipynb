–î–∏–Ω–∞–º–∏–∫–∞ —Ç–µ–ª–∞ –≤ —ç–ª–µ–∫—Ç—Ä–æ–º–∞–≥–Ω–∏—Ç–Ω–æ–º –ø–æ–¥–≤–µ—Å–µ –æ–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–µ–π

$\begin{cases}

\dot{x_{3}} = - \frac{1 + x_{3}}{1 - x_{1}}x_{2} - a(1 - x_{1})x_{3} + (1 - x_{1})u
\end{cases}$,

–≥–¥–µ –ø–∞—Ä–∞–º–µ—Ç—Ä $ùëé = 5$ –æ—Ç–≤–µ—á–∞–µ—Ç –∑–∞ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ ¬´–º–µ—Ö–∞–Ω–∏—á–µ—Å–∫–æ–π¬ª —á–∞—Å—Ç–∏ —Å–∏—Å—Ç–µ–º—ã –∫ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω–æ–º—É –≤—Ä–µ–º–µ–Ω–∏ ¬´—ç–ª–µ–∫—Ç—Ä–∏—á–µ—Å–∫–æ–π¬ª —á–∞—Å—Ç–∏. 

–î–∞–Ω–Ω–∞—è —Å–∏—Å—Ç–µ–º–∞ –∏–º–µ–µ—Ç –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–µ –Ω–µ—É—Å—Ç–æ–π—á–∏–≤–æ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ —Ä–∞–≤–Ω–æ–≤–µ—Å–∏—è –≤ —Ç–æ—á–∫–µ $(0, 0, 0)$. –õ–∏–Ω–µ–∞—Ä–∏–∑—É–µ–º –¥–∞–Ω–Ω—É—é —Å–∏—Å—Ç–µ–º—É –≤ –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ —Å–æ—Å—Ç–æ—è–Ω–∏—è —Ä–∞–≤–Ω–æ–≤–µ—Å–∏—è:

$\begin{cases}

\dot{x_{2}} = x_{1} + x_{3} \\
\dot{x_{3}} = - x_{2} - ax_{3} + u
\end{cases}$

$A = \left[\begin{matrix}0 & 1 & 0\\1 & 0 & 1\\0 & -1 & -5\end{matrix}\right]$, $B = \left[\begin{matrix}0\\0\\1\end{matrix}\right]$

—Å—á–∏—Ç–∞—è, —á—Ç–æ —É–ø—Ä–∞–≤–ª–µ–Ω–∏–µ —è–≤–ª—è–µ—Ç—Å—è –∫—É—Å–æ—á–Ω–Ω–æ –ø–æ—Å—Ç–æ—è–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–µ–π, —Ç–æ –µ—Å—Ç—å $u(t) = u_{k} = const$ –ø—Ä–∏ $t_k \leq t < t_{k + 1}$ –∏ $t_{k + 1} - t_k = h = const$.

–ò—Å—Ö–æ–¥—è –∏–∑ –≤—ã—à–µ–æ–ø–∏—Å–∞–Ω–Ω–æ–≥–æ, –ø–æ–ª—É—á–∞–µ–º:

$\xi_{k + 1} = e^{Ah}\xi{k} + \int_{t_{k}}^{t_{k+1}}e^{A(t_{k + 1} - s)}ds Bu_{k}$. 

–ë–µ—Ä–µ–º –ø–æ–ª—É—á–µ–Ω–Ω—ã–π –∏–Ω—Ç–µ–≥—Ä–∞–ª, —Å–¥–µ–ª–∞–≤ –ø–µ—Ä–µ–¥ —ç—Ç–∏–º –∑–∞–º–µ–Ω—É $v = t_{k + 1} - s$ –∏ –ø–µ—Ä–µ—Å—á–∏—Ç–∞–≤ –ø—Ä–µ–¥–µ–ª—ã –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è:

$\xi_{k + 1} = e^{Ah}\xi{k} + \int_0^he^{Av}dv Bu_{k} = e^{Ah}\xi{k} + A^{-1}(e^{Ah} - I)Bu_{k}$. –ú—ã –∑–Ω–∞–µ–º, —á—Ç–æ –ø—Ä–∏ —Ç–∞–∫–æ–π —Å—Ö–µ–º–µ –¥–∏—Å–∫—Ä–µ—Ç–∏–∑–∞—Ü–∏–∏ –≤—Å–µ —Å–≤–æ–π—Å—Ç–≤–∞ —Å–∏—Å—Ç–µ–º—ã —Å–æ—Ö—Ä–∞–Ω—è—é—Ç—Å—è.

–î–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –æ–∫–æ–Ω—á–∞—Ç–µ–ª—å–Ω–æ–≥–æ –≤–∏–¥–∞ –Ω–∞—à–µ–π –¥–∏—Å–∫—Ä–µ—Ç–Ω–æ–π —Å–∏—Å—Ç–µ–º—ã –ø–æ—Å—á–∏—Ç–∞–µ–º –º–∞—Ç—Ä–∏—Ü—ã, —Å—Ç–æ—è—â–∏–µ –ø—Ä–∏ $\xi{k}$ –∏ $u_{k}$:

–î–ª—è —ç—Ç–æ–≥–æ –Ω–∞–º –ø–æ—Ç—Ä–µ–±—É–µ—Ç—Å—è –ø–æ—Å—á–∏—Ç–∞—Ç—å –º–∞—Ç—Ä–∏—á–Ω—É—é —ç–∫—Å–ø–æ–Ω–µ–Ω—Ç—É $e^{Ah}$, –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞–≤—à–∏—Å—å —Å–ª–µ–¥—É—é—â–∏–º —Å–≤–æ–π—Å—Ç–≤–æ–º:

$e^A = Se^DS^{-1}$, –≥–¥–µ $A = Ah$, $D$ - –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ —Å —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–º–∏ —á–∏—Å–ª–∞–º–∏ –º–∞—Ç—Ä–∏—Ü—ã $Ah$, –∞ $S$ - –º–∞—Ç—Ä–∏—Ü–∞, —Å–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω–∞—è –∏–∑ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤.
